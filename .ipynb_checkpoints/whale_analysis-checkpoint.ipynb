# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
#Read risk free rate
risk_free_rate_data = Path("risk_free_rate.csv")
risk_free_rate = pd.read_csv(risk_free_rate_data, index_col="Date",infer_datetime_format=True, parse_dates=True)
risk_free_rate.sort_index()
```

```python
whale_returns_data = Path("whale_returns.csv")
whale_returns = pd.read_csv(whale_returns_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
```

```python
# Reading whale returns
whale_returns.sort_index()
```

```python
# Count nulls
whale_returns.isnull().sum()
```

```python
# Drop nulls
whale_returns.dropna(inplace=True)
whale_returns
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_returns_data = Path("algo_returns.csv")
algo_returns = pd.read_csv(algo_returns_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_returns.sort_index()
```

```python
# Count nulls
algo_returns.isnull()
```

```python
# Drop nulls
algo_returns.dropna(inplace=True)
algo_returns
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx_history = Path("sp_tsx_history.csv")
sp_closing = pd.read_csv(sp_tsx_history, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp_closing.sort_index()
sp_closing.isnull().sum()
```

```python
# Check Data Types
sp_closing.dtypes
```

```python
# Fix Data Types
sp_closing['Close']=sp_closing['Close'].str.replace(',','',regex=True)
sp_closing['Close']=sp_closing['Close'].str.replace('$','',regex=True)
sp_closing['Close']=sp_closing['Close'].astype('float')
sp_closing.dtypes
```

```python
# Calculate Daily Returns
sp_returns = sp_closing.pct_change()
sp_returns
```

```python
# Drop nulls
sp_returns.dropna(inplace=True)
sp_returns.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_returns.columns = ["SP"]
sp_returns
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
combined_returns = pd.concat([whale_returns, algo_returns,sp_returns], axis="columns", join="inner")
new_columns = ["Soros", "Paulson", "Tiger", "Berkshire", "Algo1", "Algo2", "SP"]
combined_returns.columns = new_columns
combined_returns
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the S&P TSX 60.
5. Calculate the Annualized Standard Deviation.

### Create a box plot for each portfolio


```python
# Box plot to visually show risk
box_plot = combined_returns.plot.box(rot=90)
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
soros_std = combined_returns['Soros'].std()
paulson_std = combined_returns['Paulson'].std()
tiger_std = combined_returns['Tiger'].std()
berkshire_std = combined_returns['Berkshire'].std()
algo1_std = combined_returns['Algo1'].std()
algo2_std = combined_returns['Algo2'].std()
#Output daily standard deviations of all portfolios
print(f"The Soros daily standard deviation is {soros_std}")
print(f"The Paulson daily standard deviation is {paulson_std}")
print(f"The Tiger daily standard deviation is {tiger_std}")
print(f"The Berkshire daily standard deviation is {berkshire_std}")
print(f"The Algo 1 daily standard deviation is {algo1_std}")
print(f"The Algo2 daily standard deviation is {algo2_std}")
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate  the daily standard deviation of S&P TSX 60
sp_std = combined_returns['SP'].std()
count = 0
print(f"The S&P TSX 60 daily standard deviation is {sp_std}")
# Determine which portfolios are riskier than the S&P TSX 60
portfolio_std = [soros_std, paulson_std, tiger_std, berkshire_std, algo1_std, algo2_std]
for x in portfolio_std:
    if x > sp_std:
        print(f"{new_columns[count]} is risker than S&P")
    else: print(f"{new_columns[count]} is not riskier than S&P")
    count += 1
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
soros_std_annual = soros_std*np.sqrt(252)
paulson_std_annual = paulson_std*np.sqrt(252)
tiger_std_annual = tiger_std*np.sqrt(252)
berkshire_std_annual = berkshire_std*np.sqrt(252)
algo1_std_annual = algo1_std*np.sqrt(252)
algo2_std_annual = algo2_std*np.sqrt(252)
sp_std_annual = sp_std*np.sqrt(252)
print(soros_std_annual)
print(paulson_std_annual)
print(berkshire_std_annual)
print(algo1_std_annual)
print(algo2_std_annual)
print(sp_std_annual)
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
soros_rolling_std = combined_returns['Soros'].rolling(window=21).std()
#soros_rolling_std.plot()
paulson_rolling_std = combined_returns['Paulson'].rolling(window=21).std()
#paulson_rolling_std.plot()
tiger_rolling_std = combined_returns['Tiger'].rolling(window=21).std()
#tiger_rolling_std.plot()
berkshire_rolling_std = combined_returns['Berkshire'].rolling(window=21).std()
#berkshire_rolling_std.plot()
algo1_rolling_std = combined_returns['Algo1'].rolling(window=21).std()
#algo1_rolling_std.plot()
algo2_rolling_std = combined_returns['Algo2'].rolling(window=21).std()
sp_rolling_std = combined_returns['SP'].rolling(window=21).std()
rolling_std = [soros_rolling_std,paulson_rolling_std,tiger_rolling_std, berkshire_rolling_std,algo1_rolling_std,algo2_rolling_std, sp_rolling_std]
# Plot the rolling standard deviation
ax = soros_rolling_std.plot(figsize = (20,10),title = 'Rolling 21 day standard deviation of stocks')
paulson_rolling_std.plot(ax=ax)
tiger_rolling_std.plot(ax=ax)
berkshire_rolling_std.plot(ax=ax)
algo1_rolling_std.plot(ax=ax)
algo2_rolling_std.plot(ax=ax)
sp_rolling_std.plot(ax=ax)
ax.legend(['Soros','Paulson', 'Tiger', 'Berkshire', 'Algo 1', 'Algo 2', 'SP'])
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = combined_returns.corr()
# Display de correlation matrix
print(correlation)
sns.heatmap(correlation, vmin=-1, vmax=1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate variance of S&P TSX
rolling_variance = combined_returns['SP'].rolling(window=60).var()

# Calculate covariance of a single portfolio
rolling_covariance_soros = combined_returns['Soros'].rolling(window=60).cov(combined_returns['SP'])
# Computing beta
rolling_beta_soros = rolling_covariance_soros/rolling_variance
# Plot beta trend
rolling_beta_soros.plot()
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
# Calculate the rolling standard deviation for all portfolios using a 21-day window
soros_rolling_ewmstd = combined_returns['Soros'].ewm(span=21).std()
paulson_rolling_ewmstd = combined_returns['Paulson'].ewm(span=21).std()
tiger_rolling_ewmstd = combined_returns['Tiger'].ewm(span=21).std()
berkshire_rolling_ewmstd = combined_returns['Berkshire'].ewm(span=21).std()
algo1_rolling_ewmstd = combined_returns['Algo1'].ewm(span=21).std()
algo2_rolling_ewmstd = combined_returns['Algo2'].ewm(span=21).std()
sp_rolling_ewmstd = combined_returns['SP'].ewm(span=21).std()
# Plot the rolling standard deviation
ax = soros_rolling_ewmstd.plot(figsize = (20,10), title = 'exponentially weighted 21-day rolling standard deviation')
paulson_rolling_ewmstd.plot(ax=ax)
tiger_rolling_std.plot(ax=ax)
berkshire_rolling_ewmstd.plot(ax=ax)
algo1_rolling_ewmstd.plot(ax=ax)
algo2_rolling_ewmstd.plot(ax=ax)
sp_rolling_ewmstd.plot(ax=ax)
ax.legend(['Soros','Paulson', 'Tiger', 'Berkshire', 'Algo 1', 'Algo 2', 'SP'])
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
#Assume rfr same as rfr used in week 3 class 3 exercise 10
#concatenate risk free rate with returns data
combined_returns_rfr = pd.concat([combined_returns, risk_free_rate], axis="columns", join="inner")
new_columns = ["Soros", "Paulson", "Tiger", "Berkshire", "Algo1", "Algo2", "SP","rfr"]
combined_returns_rfr.columns = new_columns
combined_returns_rfr

sharpe_ratios = ((combined_returns_rfr.mean() - combined_returns_rfr['rfr'].mean())* 252) / (combined_returns_rfr.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

### Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.

Algo 1 outperforms both the market (S&P TSX 60) and the whales portfolios, given it has the highest Sharpe ratio. Algo 2 outperforms the market (S&P TSX 60) but doesn't outperform the Berkshire whale portfolio. It outperforms the other whale portfolios.
